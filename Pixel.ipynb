# Pixel

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

# Create Filter

```python
img_col = cv2.imread('fish.jpg')
img_gray = cv2.imread('fish.jpg', cv2.IMREAD_GRAYSCALE)

plt.subplot(1,2,1), plt.imshow( cv2.cvtColor(img_col, cv2.COLOR_BGR2RGB)), plt.title('Original Color')
plt.subplot(1,2,2), plt.imshow( cv2.cvtColor(img_gray, cv2.COLOR_BGR2RGB)), plt.title('Original GRAY')
```

```python
img_4input = np.vstack((np.hstack((img_col, img_col)), 
                            np.hstack((img_col, img_col))))

plt.imshow(cv2.cvtColor(img_4input, cv2.COLOR_BGR2RGB))
```

```python
pokemon = cv2.imread("pokemon.jpg")
opencv = cv2.imread("opencv.png")
python = cv2.imread("python.png")
plt.figure(figsize=(10,20))
plt.subplot(1,3,1), plt.imshow( cv2.cvtColor(pokemon,cv2.COLOR_BGR2RGB) ), plt.title("pokemon")
plt.subplot(1,3,2), plt.imshow( cv2.cvtColor(opencv,cv2.COLOR_BGR2RGB) ), plt.title("opencv")
plt.subplot(1,3,3), plt.imshow( cv2.cvtColor(python,cv2.COLOR_BGR2RGB) ), plt.title("python")

print(pokemon.shape)
print(opencv.shape)
print(python.shape)
```

```python
mask = np.zeros_like(opencv)

condition = (opencv[...,0]==255) & \
            (opencv[...,1]==255) & \
            (opencv[...,2]==255)
mask[condition] = (255,255,255)
plt.imshow( cv2.cvtColor(mask,cv2.COLOR_BGR2RGB) )
```

```python
condition = (mask[...]==0)
pokemon[0:opencv.shape[0], 0:opencv.shape[1]][condition] = opencv[condition]
plt.imshow( cv2.cvtColor(pokemon,cv2.COLOR_BGR2RGB) )
```
